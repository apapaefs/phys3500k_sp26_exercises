*Exercise 1*: Create a three-dimensional class of vectors that contans the following: 

- A function to calculate the length of the vector.
- A function that takes the dot product of two vectors.
- A function that takes the cross product between two vectors. 

```python
import math

# define the Vector3 class, with length(), dot() and cross() functions to return:
# lenght of the vector, the dot product with another vector, and the cross product with another vector
class Vector3:
    """A class that handles 3-dimensional vectors"""
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    
    # the lenght of the vector:
    def length(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

    # take the dot product of two vectors
    def dot(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    # take the cross product of two vectors:
    def cross(self,other):
        return Vector3(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x)
```

Now let's test it! 

```python
# test the dot product
A = Vector3(1,1,1)
B = Vector3(-1,1,-1)
S = A.dot(B)
print('A.B=', S)

# test the cross product
a = Vector3(1, 0, 0)
b = Vector3(0, 1, 0)
c = a.cross(b)
print('c_x, c_y, c_z=', c.x, c.y, c.z)  # should be 0 0 1
```
