```python
import math

# class definition for a three-dimensional vector
class Vector3:
    """A class of three-dimensional vectors"""
    # initialize with x, y, and z
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    # the dot product between two vectors: returns a scalar
    def dot(self, other):
        if isinstance(other, Vector3):
            return self.x*other.x + self.y*other.y + self.z*other.z
        else:
            print(other, 'is not a of type Vector3')

    # define the cross product between two vectors:
    def cross(self, other):
        return Vector3(self.y*other.z - other.y*self.z, other.x*self.z - self.x*other.z, self.x*other.y - other.x*self.y)
```

```python
# test our Vector3 class here
a = Vector3(1.0, 1.0, 1.0)
b = Vector3(2.0, -3.0, 0)
# take the dot product
adotb = a.dot(b)
print("a.b=", adotb)

# define other vectors
A = Vector3(1,1,0)
B = Vector3(-1,1,0)
# take the cross product
C = A.cross(B)
print('C=A x B=', C.x, C.y, C.z)

# let's make an error on what we provide to the dot product:
p = Vector3(1.0, 1.0, 1.0)
r = 1.0
pdotr = p.dot(r)
```
