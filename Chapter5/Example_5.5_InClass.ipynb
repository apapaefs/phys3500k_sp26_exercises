```python
# In class solution to Example 5.5
import random
import math

# Now let's also define the function that we wish to integrate:
def f(x):
    return 1/(1+x**2)
# the real weight function: maximum occurs at x = 1 for w=3/5
def w(x): 
    return (6/5) * (1 - 0.5 * x**2) 

def mcint_vonneumann(func, weightfunc, a, b, N, wmax=None, Nscan=10000):
    """Calculates the one-dimensional Monte Carlo integral of func in [a,b] for N points"""
    # check if wmax is provided or not:
    wrunning_maximum = -1E99
    if wmax==None: 
        # scan to find wmax
        for k in range(Nscan):
            xk = (b-a)*random.random() + a
            wk = weightfunc(xk)
            if wk > wrunning_maximum:
                wrunning_maximum=wk
    else:
        pass # wmax is provided
    if wmax==None:
        wmax=wrunning_maximum
        print('Maximum found wmax=', wmax)
    # apply hit-or-miss (von Neumann):
    generatedN = 0
    xgenerated = []
    while generatedN < N:
        # pick a random number in [0,1]
        eta = random.random()
        # pick a point:
        xtrial = (b-a)*random.random() + a
        # probability to accept this point:
        paccept = weightfunc(xtrial)/wmax
        # accept or reject:
        if eta < paccept:
            xgenerated.append(xtrial)
            generatedN+=1
        
    sumf = 0 # we will use this variable for the sum of f(x_i)
    sumfsq = 0 # and this one for the sum of f(x_i)^2, used in the error calculation
    for i in range(int(N)):
        xi = xgenerated[i]
        sumf = sumf + func(xi)/weightfunc(xi)
        sumfsq = sumfsq + (func(xi)/weightfunc(xi))**2 
    # now calculate the average value of f (i.e. the integral):
    I = sumf/N
    # and the error: 
    sigmaIsq = (1/N) * ( (1/N) * sumfsq - I**2 ) # this is the variance (i.e. the error squared)
    sigmaI = math.sqrt(sigmaIsq) # this is the actual error
    return I, sigmaI # return the integral and its error

wmaximum=6/5
N=10000
# maximum provided: 
IntVN, ErrVN = mcint_vonneumann(f,w,0,1,N,wmax=wmaximum) # this way you can access both the integral and its error
# maximum NOT provided (calculated automatically):
IntVN_auto, ErrVN_auto = mcint_vonneumann(f,w,0,1,N) # this way you can access both the integral and its error
print('The estimate of the integral is=', IntVN, '+-', ErrVN)
print('The estimate of the integral is (auto maximum)=', IntVN_auto, '+-', ErrVN_auto)
print("Let's compare to the analytical value=", math.pi/4)
```
